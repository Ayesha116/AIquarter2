```python
import tensorflow as tf
```

```python
from tensorflow.keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
train_images.shape
```

```python
train_images.ndim
```

```python
train_labels
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
train_images[0]
```

```python
digit = train_images[7]
                    
import matplotlib.pyplot as plt
plt.imshow(digit,cmap=plt.cm.binary)
plt.show()
```

```python
#making a network

network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network.add(layers.Dense(10, activation='softmax'))

```

```python
train_images
```

```python
#compilation step

network.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
#chaingeing dimensions of train_images and test images because network accepts 1 dimesion images

train_images = train_images.reshape(60000,784) #reshaping
train_images = train_images.astype('float32')/255 #scaling

test_images = test_images.reshape(10000,784)
test_images = test_images.astype('float32')/255

```

```python
train_labels
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
test_labels[0]
```

```python
train_labels[0]
```

```python
 network.fit(train_images, train_labels, epochs=5, batch_size=128)
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
```
